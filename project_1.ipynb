```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
dataframe = pd.read_csv("scottish_hills.csv")

# Extract the Series for "Height" and "Latitude"
x = dataframe.Height
y = dataframe.Latitude

# Plot the height against latitude
plt.scatter(x, y)
plt.xlabel('Latitude')
plt.ylabel('Height')
plt.title('Relationship between Latitude and Height of Scottish Hills')
plt.show()

filtered_data = dataframe[dataframe['Height'] > 1000]
print(filtered_data.head())

grouped_data = dataframe.groupby('Latitude')['Height'].mean()
print(grouped_data.head())

plt.hist(dataframe['Height'], bins=50)
plt.xlabel('Height (m)')
plt.ylabel('Frequency')
plt.title('Distribution of Hill Heights')
plt.show()

correlation = dataframe['Height'].corr(dataframe['Latitude'])
print(f'Correlation between height and latitude: {correlation:.2f}')
```
